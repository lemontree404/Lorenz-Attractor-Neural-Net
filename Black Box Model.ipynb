# Black Box

```python
import keras
from keras import layers
import pandas as pd, numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
%matplotlib inline
```

```python
model = keras.Sequential()
model.add(layers.Dense(10,activation="relu"))
model.add(layers.Dense(10,activation="relu"))
model.add(layers.Dense(3,activation="linear"))
```

```python
model.compile(loss='mse',optimizer='adam')
```

```python
db = pd.read_csv("data")
db.drop("Unnamed: 0",inplace=True,axis=1)
db
x = db.iloc[:,:3]
y = db.iloc[:,-3:]

x = np.array(x).reshape(100000,3)
y = np.array(y).reshape(100000,3)
```

```python
model.fit(x,y,epochs=50,verbose=1)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
%matplotlib inline

x = 0.4
y = 0.4
z = 0.4

sigma = 10 
rho = 28
beta = 8/3

iterations=10000

lx=[x]
ly=[y]
lz=[z]

def get_array(x,y,z,lx,ly,lz):
    for i in range(iterations):
        x,y,z = model.predict(np.array([x,y,z]).reshape(1,-1))[0]
        lx+=[x]
        ly+=[y]
        lz+=[z]
        
get_array(x,y,z,lx,ly,lz)
    



fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot(lx,ly,lz,lw=0.1)
```

Model is able to reproduce a rough estimation of the Lorenz Attractor
